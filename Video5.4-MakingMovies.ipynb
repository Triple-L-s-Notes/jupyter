```python
from matplotlib.animation import FuncAnimation, FFMpegWriter
```

```python
#basic animation
nums = np.arange(0, 10, 0.1)
#print(nums)
sin, = plt.plot(nums, np.sin(nums))
frames = np.arange(0, 2*np.pi, 0.1) #overrides the default frames behavior
interval = 60 #200 ms between frames by default
def update(i): #by default, i counts up in integers from 0
    sin.set_ydata(np.sin(nums+i))
    plt.title('index: {:^2.1f}'.format(i))
#assignment of FuncAnimation to a variable is required for some reason in Thonny:
progress_callback = lambda i, n: print(f'Saving frame {i} of {n}')
animation = FuncAnimation(plt.gcf(), #get current figure or gcf
                          update, #the function to update
                          frames=frames, #overrides the default frames behavior of counting up i in integers from 0
                          interval=interval,
                          #repeat = False, #option to stop animation after one iteration
                          repeat_delay = 2e3, #optional delay in ms; 1e3 = 1000 ms = 1 s
                          #progress_callback=progress_callback
                          )

# ani = animation.FuncAnimation(fig, animate, range(1, len(y)), interval=dt*1000, blit=True, init_func=init)
# from matplotlib.animation import FFMpegWriter
writer = FFMpegWriter(fps=15, metadata=dict(artist='Stephen Shadle'), bitrate=1800)
#ani.save("/home/pi/Documents/Thonny/dynamic_image.mp4", writer=writer) #explicit directory
animation.save("sine_waveA.mp4", writer=writer) #current directory
```
